```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
file1 = 'assignment_image1.jpg'
img1 = cv2.imread(file1)
color = ('b','g','r')
plt.figure()
for i,col in enumerate(color):
    histr = cv2.calcHist([img1],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
